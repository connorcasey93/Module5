```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np 
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_mouse = pd.merge(mouse_metadata, study_results,on="Mouse ID")

# Display the data table for preview
combined_mouse.head()
```

```python
# Checking the number of mice.
mice_number = combined_mouse["Mouse ID"].nunique()
mice_number
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mouse = combined_mouse.loc[combined_mouse.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse = combined_mouse[combined_mouse['Mouse ID'].isin(duplicate_mouse)==False]
clean_mouse
```

```python
cleanmice_number = clean_mouse["Mouse ID"].nunique()
cleanmice_number
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

mean = clean_mouse['Tumor Volume (mm3)'].groupby(clean_mouse['Drug Regimen']).mean()
median = clean_mouse['Tumor Volume (mm3)'].groupby(clean_mouse['Drug Regimen']).median()
var = clean_mouse['Tumor Volume (mm3)'].groupby(clean_mouse['Drug Regimen']).var()
std = clean_mouse['Tumor Volume (mm3)'].groupby(clean_mouse['Drug Regimen']).std()
sem = clean_mouse['Tumor Volume (mm3)'].groupby(clean_mouse['Drug Regimen']).sem()

summary_mouse = pd.DataFrame({"Mean Tumor Volume":mean, "Median Tumor Volume":median,"Tumor Volume Variance":var, 
"Tumor Volume Std. Dev.":std,"Tumor Volume Std. Err.":sem})
summary_mouse
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

single_line =  clean_mouse.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
single_line
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens

# Start by getting the last (greatest) timepoint for each mouse

select_drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
select_drugs = clean_mouse[clean_mouse['Drug Regimen'].isin(select_drugs)]

# Find the last (greatest) timepoint for each mouse within the selected treatment regimens
last_time = select_drugs.groupby('Mouse ID')['Timepoint'].max()


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_volume = pd.merge(select_drugs,last_time, on=['Mouse ID', 'Timepoint'])
final_volume
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)

tumor_volume = []
# Calculate the IQR and quantitatively determine if there are any potential outliers.
    
# Locate the rows which contain mice on each drug and get the tumor volumes
 
# add subset

# Determine outliers using upper and lower bounds
for drug in treatments:
    final_tumor_volume = final_volume.loc[final_volume["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    tumor_volume.append(final_tumor_volume)
    quartiles = final_tumor_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_volume.loc[(final_tumor_volume < lower_bound) | (final_tumor_volume > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")
fig1, ax1 = plt.subplots()
plt.boxplot(tumor_volume,labels=treatments, flierprops=dict(markerfacecolor='r',))
ax1.set_ylabel('Final Tumor Volumes (mm3')
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_mouse = clean_mouse[clean_mouse["Drug Regimen"] == "Capomulin"]
mouse_capo = capomulin_mouse[capomulin_mouse["Mouse ID"] == "l509"]
mouse_capo.plot(x="Timepoint", y="Tumor Volume (mm3)")
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatmeant of mouse l509')
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_mouse = clean_mouse.loc[clean_mouse["Drug Regimen"]=="Capomulin",:].groupby("Mouse ID")[["Weight (g)","Tumor Volume (mm3)"]].mean()
plt.scatter(capomulin_mouse["Weight (g)"],capomulin_mouse["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
```

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

correlation_mouse=round(st.pearsonr(capomulin_mouse['Weight (g)'],capomulin_mouse['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {correlation_mouse}")

capo_slope, capo_int, capo_r, capo_p, capo_std_err = st.linregress(capomulin_mouse["Weight (g)"],capomulin_mouse["Tumor Volume (mm3)"])
capo_fit = capo_slope * capomulin_mouse["Weight (g)"] + capo_int
plt.scatter(capomulin_mouse["Weight (g)"],capomulin_mouse["Tumor Volume (mm3)"],color="b")
plt.plot(capomulin_mouse["Weight (g)"],capo_fit,"--",color="r")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
